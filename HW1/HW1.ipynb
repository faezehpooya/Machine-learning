```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

usjobs = pd.read_csv("usjobs.csv")
usjobs.head(5)
```

<div dir="rtl">
یه اتفاق بدی که معمولا برای دیتاست‌ها میوفته اینه که یک‌سری از داده‌ها تهی هستن و این کار معمولا تحلیل داده رو برای 
  ما سخت می‌کنه.  پس بیاید قبل از هرکاری ببینیم چقدر این داده‌ها تهی هستن.  برای این کار تعداد و درصد داده‌های تهی هر ستون رو بگید. 
(۴ امتیاز)
</div>

```python
null_count_per_column = usjobs.isnull().sum().to_frame("Null Count") 
null_count_per_column['Null Percent'] = (null_count_per_column["Null Count"]
                                         .map(lambda x: round(x / usjobs.shape[0] * 100, 2)))
null_count_per_column
```

<div dir="rtl">
    همون‌طور که می‌بینید در بعضی از ستون‌ها بیش از ۵۰ درصد داده تهی هستن. ستون‌هایی که این ویژگی رو دارن از مجموعه داده‌ها حذف کنید.
    (۴ امتیاز)
</div>

```python
columns_to_drop = null_count_per_column[null_count_per_column["Null Percent"] > 50].index
cleaned_usjobs = usjobs.drop(columns_to_drop, axis=1)
cleaned_usjobs.head()
```

<div dir="rtl">
در بعضی از ستون‌ها فقط یک داده‌ی تهی وجود داره. سطرهای مربوط به این داده‌های خالی رو از مجموعه‌ی داده‌ها حذف کنید و ابعاد مجموعه داده‌های جدید را نمایش بدید.
    (۴ امتیاز)
</div>

```python
one_null_columns = null_count_per_column[null_count_per_column["Null Count"] == 1].index
cleaned_usjobs.dropna(subset=one_null_columns, inplace=True)
cleaned_usjobs.shape
```

    در سلول زیر روش‌های متداول برای حل مشکل

<div dir="rtl">
از جمله روش‌های متداول برای حل مشکل data missing در dataset ها به شرح زیر است:
    <br>
۱. حذف سطر‌ها: <br> 
معمولا از این روش برای کنترل مقادیر null استفاده می‌شود. اگر بیش از نیمی از داده‌های یک ستون null باشد می‌توان کل آن ستون را حذف کرد. سطرهایی که یک یا چند مقدار null  دارند نیز می‌توانند حذف شوند. مانند کاری که در دو قسمت قبلی انجام دادیم. این روش فقط هنگامی توصیه می‌شود که نمونه‌های کافی در مجموعه داده‌ها وجود داشته باشد.  باید اطمینان حاصل شود که پس از حذف داده‌ها، داده‌ها بایاس نشوند. حذف داده‌ها منجر به از دست رفتن اطلاعات می‌شود که در هنگام پیش‌بینی خروجی، نتایج مورد انتظار را نخواهد داد. 
    <br>
    <br>
۲. جایگزینی با Mean/Median/Mode:
    <br>
این استراتژی می‌تواند بر روی ستون‌هایی اعمال شود که دارای داده‌های عددی است. ستون‌های مجموعه داده که دارای مقادیر عددی پیوسته هستند می‌توانند با میانگین، میانه یا حالت مقادیر باقی‌مانده در ستون جایگزین شوند. این روش می‌تواند از از دست رفتن داده‌ها در مقایسه با روش قبلی جلوگیری کند. جایگزینی دو تقریب میانگین و میانه یک روش آماری برای کنترل مقادیر از دست رفته است.
    <br>
    <br>
۳. اختصاص یک دسته خاص:
    <br>
    وقتی مقادیر از دست رفته از ستون‌هایی باشند که دسته‌بندی شده‌‌ است و تعداد مقادیر ممکن برای داده‌‌های آن‌‌ مشخص و محدود است، می‌توانیم دسته‌ی خاصی را برای مقادیر از دست رفته اختصاص دهیم. برای مثال در دیتاست این تمرین ستون‌های Rating و Size دسته‌بندی‌های مشخصی دارند. بنابراین می‌توانیم به جای مقادیر null این ستون‌ها،  دسته‌ي Unknown را قرار دهیم.  
    <br>
    <br>
۴. استفاده از الگوریتم هایی که مقادیر از دست رفته را پشتیبانی می کنند:
    <br>
KNN  یک الگوریتم یادگیری ماشین است که بر اساس اصل اندازه‌گیری فاصله کار می‌کند. می‌توان از این الگوریتم زمانی که مقادیر null  در مجموعه داده وجود دارد استفاده کرد. KNN  برای مقادیر از دست رفته به دنبال نزدیک‌ترین همسایه در بین نمونه‌های دیگر در داده‌های موجود می‌گردد.
الگوریتم دیگری که می‌تواند در اینجا مورد استفاده قرار گیرد RandomForest است. این مدل به دلیل اینکه روی داده‌های غیرخطی و دارای دسته‌بندی به خوبی کار می‌کند، نتیجه robust ایجاد می‌کند. در مجموعه داده های بزرگ، با در نظر گرفتن واریانس زیاد یا بایاس داده‌ها، با ساختار داده سازگار می‌شود و نتایج بهتری را ایجاد می‌کند.

</div>

<div dir="rtl">
تا اینجای کار فقط داشتیم داده‌هامون رو تمیز می‌کردیم. از اینجا به بعد می‌خوایم کمی روی اونها کار کنیم. یکی از 
فاکتورهای مهم توی کار کردن میزان حقوقه. پس بیاید و ستون تخمین حقوق رو به دو ستون تبدیل کنید که حداقل و حداکثر حقوق 
درخواستی در اون درج شده.
سپس ستون میانگین حقوق را از این دو ستون جدید به دست بیارید و در انتها پنج داده‌ی اول مجموعه داده رو نمایش بدید.
    (۸ امتیاز)
</div>

```python
usjobs_updated = cleaned_usjobs.copy()
usjobs_updated.insert(3, "Min Salary", 
                      usjobs_updated["Salary Estimate"].map(lambda x: x.split('-')[0].strip()))
usjobs_updated.insert(4, "Max Salary", 
                      usjobs_updated["Salary Estimate"].map(lambda x: x.split('-')[1].split('(')[0].strip()))
(
usjobs_updated.insert(5, "Avg Salary", usjobs_updated[["Min Salary", "Max Salary"]]
              .apply(lambda x: (int(x["Min Salary"][1:-1]) + int(x["Max Salary"][1:-1])) / 2, axis=1))
)
usjobs_updated.head(5)
```

<div dir="rtl">
    می‌خوایم یه آمار کلی از حقوق کارها داشته باشیم. کاری که می‌خوایم بکنید اینه که ۵٪ از کارها با بالاترین میانگین حقوق و ۵٪ از کارها با پایین‌ترین میانگین حقوق رو حذف کنید و بعدش یه آمار کلی از میانگین حقوق‌ها بدید و همچنین بگید که بنظرتون چرا ۵٪ از بالا و ۵٪ از پایین رو حذف کردیم؟ آمار کلی‌ای که می‌خوایم شامل:  mean, std, min, max می‌شه.
    (۶ امتیاز)
</div>

<div dir="rtl">
    زیرا ۵ درصد از کار‌ها با بالاترین میانگین حقوق و ۵ دصد از کارها یا پایین‌ترین میانگین حقوق جزو quantile اول و چهارم هستند و از بیشتر داده‌ها دور هستند و معمولا شامل داده‌های پرت می‌باشند. می‌توان با حذف کردن آن‌ها داده‌ها را نرمال‌تر کرد.  
</div>

```python
normalized_usjobs = usjobs_updated.drop(usjobs_updated.nlargest(
    usjobs_updated.shape[0] * 5 // 100, ["Avg Salary"]).index)
normalized_usjobs.drop(usjobs_updated.nsmallest(
    usjobs_updated.shape[0] * 5 // 100, ["Avg Salary"]).index, inplace=True)
normalized_usjobs["Avg Salary"].describe()[["max", "min", "std", "mean"]]
```

<div dir="rtl">
در ادامه میانگین حقوق پیشنهادی شرکت‌ها رو به دست بیارید (دقت کنید برای هر شرکت باید میانگین رو روی کار‌های مختلف اون شرکت حساب کنید و در نهایت به یک عدد برسید) و بعدش نمودار توزیع تخمینی رو رسم کنید.
    (۷ امتیاز)
    </div>

```python
jobs_pivot = normalized_usjobs.pivot_table(
    index='Company Name', # the rows (turned into index)
    columns='Job Title', # the column values
    values='Avg Salary', # the field(s) to processed in each group
    aggfunc=np.mean, # group operation
)

jobs_pivot["Avg Salary"] = jobs_pivot.apply(lambda x: np.mean(x), axis=1)
avg_salary_proposed_by_companies = jobs_pivot["Avg Salary"]

# avg_salary_proposed_by_companies.name = "Avg salary proposed by companies"

# # bar plot shows Avg salary proposed by companies for 100 sample company
# plt.figure(figsize=(30,10))
# sample = avg_salary_proposed_by_companies.sample(100)
# chart = sns.barplot(x=sample.index, y=sample)
# chart.set_xticklabels(chart.get_xticklabels(), rotation=90);

# dist plot shows Density curves for Avg salary proposed by companies
sns.set_style("whitegrid", {'axes.grid' : False})
sns.kdeplot(avg_salary_proposed_by_companies);
```

<div dir="rtl">
حالا یه دسته‌بندی از رتبه بندی
    (Rating)
    شرکت‌ها و میانگین حقوق اون‌ها بدید.
    در ابتدا به جدولی که در قسمت قبل برای شرکت‌ها به دست آوردید،
    ستون Rating را اضافه کنید.
    حال
    توجه کنید که توی این دسته‌بندی ما حدود 
    Rating
    رو 
    می‌خوایم در نظر بگیریم. پس برای این کار هر ۰.۵ امتیاز رو توی یک دسته قرار میدیم. یعنی شرکت‌ها با امتیاز ۰ تا ۰.۴۹ توی یه دسته و شرکت‌ها با امتیاز ۰.۵ تا ۱ توی یه دسته و ... تا شرکتها با امتیاز ۴.۵ تا ۵ توی یه دسته قرار می‌گیرن. بعد از این دسته‌بندی میانگین حقوق هر دسته رو بدست بیارید و همه این دسته‌ها رو توی یه نمودار معنادار نمایش بدید.
   (۱۰ امتیاز)
 </div>

```python
avg_salary_Rating_per_companies = (normalized_usjobs
                                    .groupby("Company Name")[["Avg Salary", "Rating"]].mean())
categorization_by_Rating = (avg_salary_Rating_per_companies
                            .groupby(pd.cut(avg_salary_Rating_per_companies["Rating"], 
                                            np.arange(0, avg_salary_Rating_per_companies["Rating"].max() + 0.6, 0.5), 
                                            right=False, include_lowest=True)).mean())
chart = sns.barplot(x=categorization_by_Rating.index, y=categorization_by_Rating["Avg Salary"])
chart.set_xticklabels(
    chart.get_xticklabels(), 
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='x-large'
);
```

<div dir="rtl">
    به جدول شرکت‌ها ستون
    Revenue
    را اضافه کنید.

.راستی حواستون به داده‌های تهی این ستون باشه و با توجه به مقادیر دیگر این ستون بهترین تصمیم رو بگیرید

```python
revenue_and_avg_salary_of_companies = avg_salary_proposed_by_companies.to_frame()
revenue_and_avg_salary_of_companies["Revenue"] = (normalized_usjobs.groupby("Company Name")["Revenue"]
                                                  .agg(lambda x : x.iloc[0]))

# revenue_and_avg_salary_of_companies.replace('Unknown / Non-Applicable', np.NAN, inplace=True)
# revenue_and_avg_salary_of_companies.dropna(subset=["Revenue"], inplace=True)
revenue_and_avg_salary_of_companies["Revenue"].fillna('Unknown / Non-Applicable', inplace=True)

categorization_by_Revenue = revenue_and_avg_salary_of_companies.groupby("Revenue").mean()

# chart = sns.barplot(x=categorization_by_Revenue.index, y=categorization_by_Revenue["Avg Salary"])
# chart.set_xticklabels(
#     chart.get_xticklabels(), 
#     rotation=45, 
#     horizontalalignment='right',
#     fontweight='light',S
#     fontsize='x-large'
# );

categorization_by_Revenue.boxplot()
plt.show()
```

<div dir="rtl">
این مرحله از کارتون تکراریه و همون کاری که توی مرحله‌ی قبل کردید رو این بار با سایز شرکت‌ها بکنید، ببینیم
می‌شه به جمع بندی از حقوق شرکت‌ها رسید؟
باز هم توجه به داده‌های تهی مهمه که اشکالی در نمایش داده‌هاتون ایجاد نکنند و برای این مرحله هم از  boxplot استفاده کنید.  
    (۵ امتیاز)

```python
size_and_avg_salary_of_companies = avg_salary_proposed_by_companies.to_frame()
size_and_avg_salary_of_companies["Size"] = (normalized_usjobs.groupby("Company Name")["Size"]
                                                  .agg(lambda x : x.iloc[0]))

# size_and_avg_salary_of_companies.replace('Unknown', np.NAN, inplace=True)
# size_and_avg_salary_of_companies.dropna(subset=["Size"], inplace=True)
size_and_avg_salary_of_companies["Size"].fillna('Unknown', inplace=True)

categorization_by_Size = size_and_avg_salary_of_companies.groupby("Size").mean()


categorization_by_Size.boxplot()
plt.show()
```

<div dir="rtl">
    توی هر سه مرحله‌ی قبل همبستگی رو حساب کنید و ببینید که آیا این هیچ ارتباطی بین میزان حقوق و آیتم‌های معرفی شده وجود داره؟
(۵ امتیاز)    
</div>

```python
print("Correlation between Rating and Avg Salary is:", round(categorization_by_Rating.corr()["Rating"]
                                                             .loc["Avg Salary"], 6))

Revenue_and_Avg_Salary_correlation = categorization_by_Revenue.reset_index()
# Revenue_and_Avg_Salary_correlation["Revenue"] = categorization_by_Revenue.reset_index()["Revenue"].factorize()[0]

revenue_X = {'$1 to $2 billion (USD)': 1500, '$1 to $5 million (USD)': 3, 
             '$10 to $25 million (USD)' : 17.5,  '$10+ billion (USD)': 12000,
             '$100 to $500 million (USD)': 300, '$2 to $5 billion (USD)': 3500, 
             '$25 to $50 million (USD)': 37.5, '$5 to $10 billion (USD)': 7500, 
             '$5 to $10 million (USD)': 7.5, '$50 to $100 million (USD)': 75, 
             '$500 million to $1 billion (USD)': 750, 'Less than $1 million (USD)': 0.5}
revenue_X['Unknown / Non-Applicable'] = np.median(list(revenue_X.values()))

Revenue_and_Avg_Salary_correlation["Revenue"] = (Revenue_and_Avg_Salary_correlation["Revenue"]
                                                 .apply(lambda x: revenue_X[x]))

print("Correlation between Revenue and Avg Salary is:", round(Revenue_and_Avg_Salary_correlation.corr()["Revenue"]
                                                              .loc["Avg Salary"], 6))

Size_and_Avg_Salary_correlation = categorization_by_Size.reset_index()
# Size_and_Avg_Salary_correlation["Size"] = categorization_by_Size.reset_index()["Size"].factorize()[0]

size_X = {'1 to 50 employees': 25, '10000+ employees': 11000, '1001 to 5000 employees': 2500,
          '201 to 500 employees': 350, '5001 to 10000 employees': 7500, '501 to 1000 employees': 750, 
          '51 to 200 employees': 125}
size_X['Unknown'] = int(np.median(list(size_X.values())))

Size_and_Avg_Salary_correlation["Size"] = (Size_and_Avg_Salary_correlation["Size"]
                                              .apply(lambda x: size_X[x]))
print("Correlation between Size and Avg Salary is:", round(Size_and_Avg_Salary_correlation.corr()["Size"]
                                                           .loc["Avg Salary"], 6))
```
